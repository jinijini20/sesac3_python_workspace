```python
student={}
while True:
    name=input("이름을 입력하세요")
    if name == "끝":
        break
    kor,math,eng,science=map(int,input("국어 수학 영어 과학 점수를 차례대로 띄어서 입력하세요").split())
    score=[kor,math,eng,science]
    student[name]=score

choice=input("누구의 정보를 보고싶어요?")

def student_add(**kwargs): 
    avg=(sum(student[choice]))/4
    sum_num=0
    sum_len=0
    for i in student.values():
        for j in i:
            sum_num+=j
            sum_len+=1

    total_avg=(sum_num/sum_len)       
            
    
    max_sub=max(student[choice])
    min_sub=min(student[choice])
    return f"{choice}학생 정보: 학생의 따른 평균은 {avg}이고 전체평균은 {total_avg} 과목별 최대점수는 {max_sub}이고 과목별 최소점수는 {min_sub}입니다."

    
        
    
student_add(student=choice)
```
